```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Constants
g = 9.8  # gravity (m/s^2)
angles = [30, 45, 60]  # launch angles in degrees
v0 = 50  # initial velocity in m/s

# Container for all simulations
trajectories = []

# Simulate for each angle
for angle in angles:
    theta = np.radians(angle)
    t_flight = 2 * v0 * np.sin(theta) / g
    t = np.linspace(0, t_flight, num=100)

    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t**2

    df = pd.DataFrame({
        'time': t,
        'x': x,
        'y': y,
        'angle': angle,
        'velocity': v0
    })
    trajectories.append(df)

# Combine all trajectories
df_all = pd.concat(trajectories, ignore_index=True)

# Plot Trajectories
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

for angle in angles:
    subset = df_all[df_all['angle'] == angle]
    plt.plot(subset['x'], subset['y'], label=f'{angle}°')

plt.title("Projectile Motion for Different Angles")
plt.xlabel("Distance (x) [m]")
plt.ylabel("Height (y) [m]")
plt.legend()
plt.tight_layout()
plt.show()
```
